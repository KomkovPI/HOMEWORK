Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('https://lectures.scientific-python.org/_downloads/087d8fe936fcb7a4762b9850b932cf9e/populations.txt')

years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]

hares_larger = hares > lynxes

species_larger = np.where(hares_larger, 'H', 'L')

result_list = species_larger.tolist()
print("Годы с наибольшей популяцией по видам:")
print(result_list)

h_count = np.sum(hares_larger)
l_count = len(years) - h_count


plt.figure(figsize=(7, 5))

sizes = [h_count, l_count]
labels = [f'Hares ({h_count} years)', f'Lynxes ({l_count} years)']
colors = ['blue', 'red']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Ratio of Years with Largest Population: Hares vs Lynxes')
plt.show()
```

Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('https://lectures.scientific-python.org/_downloads/087d8fe936fcb7a4762b9850b932cf9e/populations.txt')

years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]

hare_changes = np.diff(hares)
lynx_changes = np.diff(lynxes)

change_years = years[1:]

correlation_coefficient = np.corrcoef(hare_changes, lynx_changes)[0, 1]

print(f"Коэффициент корреляции между изменениями популяций зайцев и рысей: {correlation_coefficient:.3f}")

plt.figure(figsize=(10, 5))

plt.plot(change_years, hare_changes, label='Change in Hare Population', color='blue')

plt.plot(change_years, lynx_changes, label='Change in Lynx Population', color='red')

plt.xlabel('Year')
plt.ylabel('Population Change')
plt.title('Yearly Changes in Hare and Lynx Populations\n' + f'Correlation coefficient: {correlation_coefficient:.3f}')
plt.grid(True)
plt.legend()
plt.show()
```

Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value with the true value and print a ratio of the error to the true value.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('https://lectures.scientific-python.org/_downloads/087d8fe936fcb7a4762b9850b932cf9e/populations.txt')

years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]

true_1920_hare = hares[-1]

recent_years = years[-5:-1]
recent_hares = hares[-5:-1]

changes = np.diff(recent_hares)

avg_change = np.mean(changes)

predicted_1920 = recent_hares[-1] + avg_change

error = predicted_1920 - true_1920_hare
relative_error = error / true_1920_hare

print("Оценка популяции зайцев в 1920 году:")
print(f"Истинное значение: {true_1920_hare:.0f}")
print(f"Прогнозируемое значение: {predicted_1920:.0f}")
print(f"Абсолютная ошибка: {error:.0f}")
print(f"Относительная ошибка: {relative_error:.3f} ({relative_error*100:.1f}%)")

plt.figure(figsize=(12, 6))

plt.plot(years[:-1], hares[:-1], 'bo-', label='Известные данные', markersize=6)
plt.plot(1920, true_1920_hare, 'go', markersize=10, label='Истинное значение 1920')
plt.plot(1920, predicted_1920, 'ro', markersize=10, label='Прогнозируемое значение 1920')
plt.plot(recent_years, recent_hares, 'ro-', markersize=8, label='Годы, используемые для прогноза')

plt.xlabel('Год')
plt.ylabel('Популяция зайцев')
plt.title('Прогнозирование популяции зайцев в 1920 году\n')
plt.legend()
plt.grid(True)

plt.show()
```

Напишите функцию, реализующую алгоритм k-means. Выделите 10 кластеров в датасете MNIST. Дайте интерпретацию каждому полученному кластеру (к какой цифре его можно отнести). Нарисуйте изображения, которые отвечают центрам кластеров. Посчитайте, сколько из точек кластера действительно являются той цифрой, к которой вы его отнесли (для этого сравните с массивом правильных ответов labels) и нарисуйте матрицу ошибок.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.metrics import confusion_matrix

def k_means(X, k, max_iter=100, threshold=1e-4):
    n_samples, n_features = X.shape
    centroids = X[np.random.choice(n_samples, k, replace=False)]
    
    for iteration in range(max_iter):
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        
        new_centroids = np.zeros_like(centroids)
        for i in range(k):
            cluster_points = X[labels == i]
            if len(cluster_points) > 0:
                new_centroids[i] = cluster_points.mean(axis=0)
            else:
                new_centroids[i] = centroids[i]
        
        if np.linalg.norm(centroids - new_centroids) < threshold:
            break
            
        centroids = new_centroids
    
    sse = 0
    for i in range(k):
        cluster_points = X[labels == i]
        if len(cluster_points) > 0:
            sse += np.sum(np.linalg.norm(cluster_points - centroids[i], axis=1)**2)
    
    return labels, sse, centroids

images, labels_true = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False)
labels_true = labels_true.astype(int)
images = images / 255.0

sample_indices = np.random.choice(len(images), 5000, replace=False)
images_sample = images[sample_indices]
labels_true_sample = labels_true[sample_indices]

labels_pred, sse, centroids = k_means(images_sample, k=10, max_iter=100, threshold=1e-4)

cluster_to_digit = {}
for i in range(10):
    mask = labels_pred == i
    if np.sum(mask) > 0:
        values, counts = np.unique(labels_true_sample[mask], return_counts=True)
        digit = values[np.argmax(counts)]
        cluster_to_digit[i] = digit

plt.figure(figsize=(12, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(centroids[i].reshape(28, 28), cmap='gray')
    digit = cluster_to_digit.get(i, '?')
    plt.title(f'{digit}')
    plt.axis('off')
plt.show()

y_pred = np.array([cluster_to_digit[label] for label in labels_pred])
cm = confusion_matrix(labels_true_sample, y_pred)

plt.figure(figsize=(8, 6))
plt.imshow(cm, cmap='Blues')
for i in range(10):
    for j in range(10):
        plt.text(j, i, str(cm[i, j]), ha='center', va='center', 
                 color='white' if cm[i, j] > cm.max()/2 else 'black')
plt.xticks(range(10))
plt.yticks(range(10))
plt.show()

correct = 0
for i in range(10):
    if i in cluster_to_digit:
        mask = labels_pred == i
        correct += np.sum(labels_true_sample[mask] == cluster_to_digit[i])

accuracy = correct / len(labels_true_sample)
print(f"Точность: {accuracy*100:.2f}%")
```
